```python
import pandas as pd
data1 = pd.read_csv('FIFA World Cup Attendance.csv')
data1
```

```python
del data1['Unnamed: 0']
```

```python
data1
```

```python
# which country has hosted how many times
data1_copy1 = data1
data1_copy1 = data1_copy1.drop(22)
data1_copy1.Hosts.value_counts()
```

```python
## top 10 hosts according to count
hosts = data1_copy1.Hosts.value_counts()
TOP_10_HOSTS_ACCORDING_TO_COUNT = hosts.head(10).plot(autopct='%1.0f%%', kind='pie',figsize=(30,7),title="TOP 10 HOSTS")
```

```python
### TOP 10 Total_Attendance in any year
data1_copy2 = data1
data1_copy2= data1_copy2.drop(22)
total_attendance1=data1_copy2.groupby(['Year'])['Total_Attendance'].max().reset_index().sort_values('Total_Attendance',ascending=False)
total_attendance1
```

```python
#Total attendance in millions
total_attendance2 =data1_copy2.groupby(['Year'])['Total_Attendance'].max().reset_index().sort_values('Year',ascending=True)
total_attendance2.plot(x='Year',y="Total_Attendance",figsize=(15,7),title="Total Attendance in Millions over the years")
```

```python
#TOP 10 least Number of matches played in any year
data1_copy3 = data1
data1_copy3= data1_copy3.drop(22)
number_of_matches1 =data1_copy2.groupby(['Year'])['Matches'].min().reset_index().sort_values('Matches',ascending=True).head(10)
number_of_matches1
```

```python
#Total number of matches
c = ['red','orange','yellow','green','black','blue']
number_of_matches2 =data1_copy2.groupby(['Year'])['Matches'].max().reset_index().sort_values('Year',ascending=True)
number_of_matches2.plot(x='Year',y="Matches",figsize=(15,7),title="Total Number of Matches Played over the years",kind="bar",color=c)
```

```python
#Top 10 Average Attendance in any year
data1_copy4= data1
data1_copy4= data1_copy4.drop(22)
avg_number_of_attendance1 =data1_copy4.groupby(['Year'])['Average_Attendance'].max().reset_index().sort_values('Average_Attendance',ascending=False).head(10)
avg_number_of_attendance1 
```

```python
#Average attendance
number_of_matches2 =data1_copy2.groupby(['Year'])['Average_Attendance'].max().reset_index().sort_values('Year',ascending=True)
number_of_matches2.plot(x='Year',y="Average_Attendance",figsize=(15,7),title="Average number of Attendance over the years")
```
